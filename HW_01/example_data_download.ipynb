```python
# Asset data download
# CQF final project: Portfolio optimization
import yfinance as yf
import pandas as pd 
import time
import matplotlib.pyplot as plt  
%matplotlib inline
%pylab inline
```

```python
data = yf.download("FB KO AAPL MCD GE ^VLIC", start="2013-01-01", end="2019-05-08")
data.fillna(method='bfill', inplace=True)
```

```python
data.columns
```

An example of historical data is shown as follows.

```python
data.head()
```

Historical data of all assets and S&P500 index are plotted as follows.

```python
pylab.rcParams['figure.figsize'] = (15, 9)   
data["Adj Close"].plot(grid = True)
```

Notice that the price range of each asset are quite different. In order to observe price trends clearer, the data of each asset was normalized by dividing every price point with the initial price at $t=0$. 

```python
asset_return = data.apply(lambda x: x / x[0])
asset_return["Adj Close"].plot(grid = True).axhline(y = 1, color = "black", lw = 2)
```

With this new scale, we can see that some assets increases along with the market index, while some assets decreases and do not look good for investment.<br/>
